# Loan Eligibility Prediction using XGBoost

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from xgboost import XGBClassifier, XGBRegressor
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

## Loading the Improved Dataset

```python

# Load the updated dataset
file_path = "Updated_Loan_Eligibility_Dataset_Improved.csv"
df = pd.read_csv(file_path)

# Display first few rows
df.head()
```

## Data Preprocessing and Feature Engineering

```python

# Separating features and target variables
X = df.drop(columns=["Loan_Eligibility", "Interest_Rate", "Loan_Tenure"])
y1 = df["Loan_Eligibility"]
y2 = df["Interest_Rate"]
y3 = df["Loan_Tenure"]

# Splitting data into train and test sets
X_train, X_test, y1_train, y1_test = train_test_split(X, y1, test_size=0.2, random_state=42)
X_train, X_test, y2_train, y2_test = train_test_split(X, y2, test_size=0.2, random_state=42)
X_train, X_test, y3_train, y3_test = train_test_split(X, y3, test_size=0.2, random_state=42)

# Display dataset shape
X_train.shape, X_test.shape
```

## Exploratory Data Analysis and Visualization

```python

# Plot Income vs Loan Eligibility
plt.figure(figsize=(8, 5))
sns.boxplot(x=y1, y=np.expm1(X["Income"]), data=df)
plt.title("Income vs Loan Eligibility")
plt.xlabel("Loan Eligibility")
plt.ylabel("Income (Original Scale)")
plt.savefig("income_vs_eligibility.png")
plt.show()

# Plot Income vs Interest Rate
plt.figure(figsize=(8, 5))
sns.scatterplot(x=np.expm1(X["Income"]), y=y2, data=df)
plt.title("Income vs Interest Rate")
plt.xlabel("Income (Original Scale)")
plt.ylabel("Interest Rate")
plt.savefig("income_vs_interest_rate.png")
plt.show()

# Detecting outliers in Income using Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(y=np.expm1(X["Income"]))
plt.title("Outlier Detection in Income")
plt.savefig("income_outliers.png")
plt.show()
```

## Training XGBoost Models

```python

# Loan Eligibility Model (Classification)
eligibility_model = XGBClassifier()
eligibility_model.fit(X_train, y1_train)

# Interest Rate Prediction Model (Regression)
interest_model = XGBRegressor()
interest_model.fit(X_train, y2_train)

# Loan Tenure Prediction Model (Regression)
tenure_model = XGBRegressor()
tenure_model.fit(X_train, y3_train)
```

## Model Evaluation

```python

# Predicting on Test Set
y1_pred = eligibility_model.predict(X_test)
y2_pred = interest_model.predict(X_test)
y3_pred = tenure_model.predict(X_test)

# Loan Eligibility Model Evaluation
accuracy = accuracy_score(y1_test, y1_pred)
print(f"Loan Eligibility Model Accuracy: {accuracy * 100:.2f}%")

# Interest Rate Model Evaluation
mae_interest = mean_absolute_error(y2_test, y2_pred)
mse_interest = mean_squared_error(y2_test, y2_pred)
print(f"Interest Rate Model MAE: {mae_interest:.2f}, MSE: {mse_interest:.2f}")

# Loan Tenure Model Evaluation
mae_tenure = mean_absolute_error(y3_test, y3_pred)
mse_tenure = mean_squared_error(y3_test, y3_pred)
print(f"Loan Tenure Model MAE: {mae_tenure:.2f}, MSE: {mse_tenure:.2f}")
```

## Predicting an Example Loan

```python

# Example input for prediction (random row from dataset)
example_input = X_test.iloc[0].values.reshape(1, -1)

# Predicting outcomes
eligibility_prediction = eligibility_model.predict(example_input)
interest_prediction = interest_model.predict(example_input)
tenure_prediction = tenure_model.predict(example_input)

print(f"Predicted Loan Eligibility: {'Eligible' if eligibility_prediction[0] == 1 else 'Not Eligible'}")
print(f"Predicted Interest Rate: {interest_prediction[0]:.2f}%")
print(f"Predicted Loan Tenure: {tenure_prediction[0]:.2f} years")
```
